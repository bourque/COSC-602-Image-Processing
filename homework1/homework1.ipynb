```python
import matplotlib.pyplot as plt

import numpy as np

from skimage.color import rgb2gray
from skimage.filters import threshold_otsu
from skimage.io import imread
from skimage.io import imsave
```

```python
image = imread('bourque.jpg')
```

```python
fig = plt.figure()
ax = plt.subplot(111)
ax.imshow(image)
plt.show()
```

```python
image_grayscale = rgb2gray(image)
```

```python
fig = plt.figure()
ax = plt.subplot(111)
ax.imshow(image_grayscale, cmap='gray')
plt.show()
imsave('bourque_grayscale.jpg', image_grayscale)
```

```python
thresh = threshold_otsu(image_grayscale)
```

```python
image_binary = np.zeros((image.shape))
image_binary[np.where(image_grayscale > thresh)] = 1.0
```

```python
fig = plt.figure()
ax = plt.subplot(111)
ax.imshow(image_binary, cmap=plt.cm.gray)
plt.show()
```

```python
imsave('bourque_binary.jpg', image_binary)
```
